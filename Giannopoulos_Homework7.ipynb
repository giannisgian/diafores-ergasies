## 4.  Τώρα που έχετε μια λίστα με τα links, αρχίστε το scraping των reports κάθε μήνα. Μπορείτε να χρησιμοποιήσετε ή beautiful soup ή selenium ή συνδυασμό των 2.

Δείτε τι περιλαμβάνει κάθε εγγραφή π.χ. εδώ: http://www.nuforc.org/webreports/ndxe201904.html

Πρέπει για κάθε εγγραφή να πάρετε Date, City, State, Shape και Summary.

```python
html = urlopen('http://www.nuforc.org/webreports/ndxe201904.html').read()
soup = BeautifulSoup(html, "html.parser")
list_data = []
for row in soup.find_all('tr')[1:]:
  tds = row.find_all('td')
  list_data.append({
    'Date': tds[0].text,
    'City': tds[1].text,
    'State': tds[2].text,
    'Shape': tds[4].text,
    'Country': tds[3].text,
    'Summary': tds[6].text
  })
list_data
```

## 6. Φτιάξτε ένα κόκκινο bar chart, σορταρισμένο στο οποίο θα εμφανίζονται τα 5 δημοφιλέστερα σχήματα των UFO

```python
df2 = pd.DataFrame (list_data)
df2['Shape'].replace('', np.nan, inplace=True)
df2.dropna(subset=['Shape'], inplace=True)


bar = df2['Shape'].value_counts().sort_values(ascending=False)[:5].plot(kind='barh', color='red')
bar.invert_yaxis()
```

## 8. Μήπως εμφανίζονται άλλα UFO ανάλογα με την περιοχή? Δείτε ποια σχήματα εμφανίζονται συχνότερα σε ποια περιοχή και φτιάξτε ένα γράφημα.

```python
states = []

for state in df['State']:
  if state not in states:
    if(len(state) == 2):
      states.append(state)


for state in states[20:40]:
  # print(state)
  print("\n")
  temp = df[df['State'] == state]
  plt.figure()
  temp['Shape'].value_counts().sort_values(ascending=False).head(10).plot(kind='barh', color='red', title=state)
  plt.show()
```

```python
groups = df.groupby(
    'Country'
    )['Shape'].value_counts(
    ).sort_values(
        ascending=False
    ).reset_index(
        name='count'
    )
groups = groups.drop_duplicates(subset=['Country'], keep='first')
groups["Label"] = groups["Country"] + "-" + groups["Shape"]
groups.plot(x="Label", kind="bar")

```
